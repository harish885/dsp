```python
def is_even(a):
    """
    this function returns a True if the given number is even,
    or it returns False
    """
    result = a % 2
    if result == 1:
        return False
    else:
        return True
```

```python
is_even(10)
```

```python
def is_even_2(a):
    """
    this function returns a True if the given number is even,
    or it returns False
    """
    if a % 2 == 1:
        return False
    else:
        return True
```

```python
is_even_2(25)
```

```python
def is_even_3(a):
    """
    this function returns a True if the given number is even,
    or it returns False
    """
    return (a % 2 == 0) # returns true if the result is zero or otherwise false, because it is a boolean
```

```python
is_even_3(1)
```
