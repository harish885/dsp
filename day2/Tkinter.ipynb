```python
import tkinter as tk
```

```python
root = tk.Tk()
```

```python
root.title("my First kinter window")
```

```python
welcome = tk.Label(root, text = "Welcome to Tkinter")
```

```python
welcome.pack()
```

```python
def on_click():
    welcome.config(text = "Button Clicked!")
```

```python
click_me_button = tk.Button(root, text = "Click me", command = on_click)
```

```python
click_me_button.pack()
```
