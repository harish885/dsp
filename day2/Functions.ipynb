# Functions

```python
def greet(name):
    name = name.upper()
    print(f"Hello {name}")
```

```python
greet("Harish")
```

```python
def greet2(name, message="Hi"):
    print(f"{message} {name}")
```

```python
greet2("Harish")
```

```python
def sum(a,b):
    return a + b
result = sum(5,6)
```

```python
def calculate_area(width,height):
    area = width * height
    return area
```

```python
calculate_area(10,25)
```

```python
def introduce(name, age = 25):
    print(f"My name is {name} and i am {age}")
```

```python
introduce("Ravi")
```

```python
def first_last(mylist):
    print(f"first element is {mylist[0]} and last element is {mylist[-1]}")
```

```python
mylist = [25,36,"Banana"]
```

```python
first_last(mylist)
```
