# WHILE LOOPS

```python
counter = 2
while counter < 5:
    print(counter)
    counter += 1
```

```python
counter = 0
while counter < 5:
    print(f"counter in loop {counter}")
    counter += 1
    if counter == 3:
        break

print(f"counter out of loop {counter}")
```
